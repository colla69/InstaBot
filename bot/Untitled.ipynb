```python
import threading
from apscheduler.schedulers.background import BlockingScheduler
from psw import psw
from user import User
from time import sleep
from datetime import datetime

scheduler = BlockingScheduler()
job = None
user = User(psw()[0], psw()[1])


def do_follow_following_followers():
    #while True:
    user.follow_following_followers()

def do_follow_followers():
    #while True:
    #    sleep(200)
    user.follow_followers()

def work():
    user.logger.log("starting work... ")
    user.login()
    start = datetime.now()
    
    threads = [
               threading.Thread(target=do_follow_following_followers),
               threading.Thread(target=do_follow_followers),
              ]
    
    [x.start() for x in threads]
    [x.join() for x in threads] # Wait for all of them to finish    
    
    #for x in threads:
    #    x.join()    
    #user.like_following()

    end = datetime.now()
    user.logger.log("duration ", end="")
    user.logger.log(end-start)


def start_job():
    global job
    scheduler.add_job(work, 'interval', seconds=18000)    
    work()
    try:
        scheduler.start()
    except Exception as e:
        print(e)


start_job()



```
